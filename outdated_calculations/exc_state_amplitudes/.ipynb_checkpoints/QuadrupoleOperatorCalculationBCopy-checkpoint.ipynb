# Quadrupole Operator calculation

The general form of a state in this model is

$$\Bigg|\hspace{1mm}\Big([f]_{\pi-}\otimes[f]_{\pi+}\Big)\otimes\Big([f]_{\nu-}\otimes[f]_{\nu+}\Big),\hspace{1mm}\Big((\lambda_{\pi-},\mu_{\pi-})\otimes(\lambda_{\pi+},\mu_{\pi+})\Big)\otimes\Big((\lambda_{\nu-},\mu_{\nu-})\otimes(\lambda_{\nu+},\mu_{\nu+})\Big),\\\hspace{1mm}(\lambda_{\pi},\mu_{\pi})\otimes (\lambda_{\nu},\mu_{\nu}),\hspace{1mm}\rho(\lambda,\mu),K,L;\hspace{1mm}J=L,M\Bigg> $$

The excited state has two contributions from the excitations of proton and neutron to their corresponding contiguous upper shells as

$$|\Phi(J^{\Pi})_{ex.}\rangle = \sqrt{\beta}|\Psi_{ex.}^{\pi}(J^{\Pi})\rangle + \sqrt{1-\beta}|\Psi_{ex.}^{\nu}(J^{\Pi})\rangle$$

where

$$|\Psi_{ex.}^{\pi}(J^{\Pi})\rangle=\Bigg|\hspace{1mm}\Big([2^{I-1},1]_{\pi}\otimes[1]_{\pi}\Big)\otimes\Big([2^{I}]_{\nu-}\otimes [0]_{\nu+}\Big),\hspace{1mm}\Big((\lambda_{\pi-},\mu_{\pi-})\otimes(\lambda_{\pi+},\mu_{\pi+})\Big)\otimes\Big((\lambda_{\nu-},\mu_{\nu-})\otimes(0,0)_{\nu+}\Big),\hspace{1mm}\\(\lambda_{\pi},\mu_{\pi})\otimes (\lambda_{\nu},\mu_{\nu}),\hspace{1mm}\rho(\lambda,\mu),K,L;\hspace{1mm}J=L,M\Bigg> $$

and 

$$|\Psi_{ex.}^{\nu}(J^{\Pi})\rangle=\Bigg|\hspace{1mm}\Big([2^{I}]_{\pi-}\otimes[0]_{\pi+}\Big)\otimes\Big([2^{I-1},1]_{\nu}\otimes [1]_{\nu}\Big),\hspace{1mm}\Big(
(\lambda_{\pi-},\mu_{\pi-})\otimes(0,0)_{\pi+}\Big)\otimes\Big((\lambda_{\nu-},\mu_{\nu-})\otimes(\lambda_{\nu+},\mu_{\nu+})\Big),\\\hspace{1mm}(\lambda_{\pi},\mu_{\pi})\otimes (\lambda_{\nu},\mu_{\nu}),\hspace{1mm}\rho(\lambda,\mu),K,L;\hspace{1mm}J=L,M\Bigg> $$

The amplitude $\beta$ must be determined. A method is shown here.


## Quadrupole operator 

The total quadrupole operator is 

$$Q^{2}_{\mu} = Q^{2}_{\mu, \pi} + Q^{2}_{\mu, \nu} =  Q^{2}_{\mu, \pi-}+  Q^{2}_{\mu, \pi+} + Q^{2}_{\mu, \nu-} + Q^{2}_{\mu, \nu+}$$


Then, the total quadrupole transition operator is given by

$$
T^{2}_{\mu}(E)= b^2_0\Big(e_{\pi}Q^{2}_{\mu, \pi-}+  e_{\pi}Q^{2}_{\mu, \pi+} + e_{\nu}Q^{2}_{\mu, \nu-} + e_{\nu}Q^{2}_{\mu, \nu+}\Big) = b^2_0\sum_{\substack{\sigma=\pi,\nu\\s=+,-}}e_{\sigma}Q^{2}_{\mu, \sigma s}
$$

its corresponding reduced matrix element (r.m.e) is

$$
\langle\Phi(J^{\Pi})_{ex.}||T^{2}(E)||\Phi(J^{\Pi})_{ex.}\rangle= b^2_0\sum_{\substack{\sigma=\pi,\nu\\s=+,-}}e_{\sigma}\Bigg(  
\beta\langle\Psi_{ex.}^{\pi}(J^{\Pi}) || Q^{2}_{\sigma s} || \Psi_{ex.}^{\pi}(J^{\Pi})\rangle
+(1-\beta)\langle\Psi_{ex.}^{\nu}(J^{\Pi})|| Q^{2}_{\sigma s} ||\Psi_{ex.}^{\nu}(J^{\Pi})\rangle
\Bigg)
$$

The quadrupole reduced matrix element is

$$\Bigg\langle\hspace{1mm}\Big([f]_{\pi-}\otimes[f]_{\pi+}\Big)\otimes\Big([f]_{\nu-}\otimes[f]_{\nu+}\Big),\hspace{1mm}\Big((\lambda_{\pi-},\mu_{\pi-})\otimes(\lambda_{\pi+},\mu_{\pi+})\Big)\otimes\Big((\lambda_{\nu-},\mu_{\nu-})\otimes(\lambda_{\nu+},\mu_{\nu+})\Big),\\\hspace{1mm}(\lambda_{\pi},\mu_{\pi})\otimes (\lambda_{\nu},\mu_{\nu}),\hspace{1mm}\rho(\lambda,\mu),K,L';\hspace{1mm}J'=L'\Bigg|\Bigg| Q^{(11),2}_{\sigma s} 
\Bigg|\Bigg|\hspace{1mm}\Big([f]_{\pi-}\otimes[f]_{\pi+}\Big)\otimes\Big([f]_{\nu-}\otimes[f]_{\nu+}\Big),\hspace{1mm}\Big((\lambda_{\pi-},\mu_{\pi-})\otimes(\lambda_{\pi+},\mu_{\pi+})\Big)\otimes\Big((\lambda_{\nu-},\mu_{\nu-})\otimes(\lambda_{\nu+},\mu_{\nu+})\Big),\hspace{1mm}(\lambda_{\pi},\mu_{\pi})\otimes (\lambda_{\nu},\mu_{\nu}),\hspace{1mm}\rho(\lambda,\mu),K,L;\hspace{1mm}J=L\Bigg\rangle\\
= \sum_{\substack{\bar{\rho}\rho_{\pi}\rho_{\nu}}}\langle(\lambda,\mu)KL; (11)2||(\lambda,\mu)KL'\rangle_{\bar{\rho}}
\begin{Bmatrix}
(\lambda_{\pi},\mu_{\pi}) & (x,y) & (\lambda_{\pi},\mu_{\pi}) & \rho_\pi\\
(\lambda_{\nu},\mu_{\nu}) & (x',y') & (\lambda_{\nu},\mu_{\nu}) & \rho_\nu\\
(\lambda,\mu) & (1,1) & (\lambda,\mu) & \bar{\rho}\\
1 & 1 & 1 &
\end{Bmatrix}
\\\times\langle (\lambda_{\sigma-},\mu_{\sigma-})\otimes(\lambda_{\sigma+},\mu_{\sigma+}), (\lambda_{\sigma},\mu_{\sigma}), S_{\sigma-}\otimes S_{\sigma+},  S_{\sigma}|||Q^{(11)}_{\sigma s} |||(\lambda_{\sigma-},\mu_{\sigma-})\otimes(\lambda_{\sigma+},\mu_{\sigma+}), (\lambda_{\sigma},\mu_{\sigma}), S_{\sigma-}\otimes S_{\sigma+},  S_{\sigma}\rangle
$$

```python
import numpy as np
import pandas as pd
```

```python
# Order 2 casimir of SU(3)
def C2(l,m):
    return l**2 + m**2 + l*m + 3.0*(l+m)

# Effective charges
def effecN(L,Z,A):
    return Z*(-1/A)**L

def effecP(L,Z,A):
    return ((A-1)**L+(-1)**L*(Z-1))/A**L
```

```python
# Function to upload SU(3) Wigner coefficients from the program's output format
# path is the direction of the file

def load_SU3_Wigner_file(path):
    
    # Filename depending on irrep
    filename = path 

    # Open file
    with open(filename + '.txt') as f:
        lines = f.readlines()
        
    # SU3 irreps loaded
    l1, m1 = int(lines[2].split(" ")[0]), int(lines[2].split(" ")[1]) 
    l2, m2 = int(lines[6].split(" ")[0]), int(lines[6].split(" ")[1]) 
    l12, m12 = int(lines[10].split(" ")[0]), int(lines[10].split(" ")[1]) 
    
    # Values and labels stored
    values = [list(filter(lambda x: x != '', l.split(" "))) for l in lines[15:]]
    
    # Column names and multiplicidades
    cols = ["k1", "L1", "k2", "L2", "k3", "L3"]
    rhos = []
    for i in range(1,len(values[0])-5):
        cols.append("rho{0}".format(i-1))
        rhos.append("rho{0}".format(i-1))
        
    # Values stored
    values_df = pd.DataFrame(np.array(values), columns = cols)
    values_df[["k1", "L1", "k2", "L2", "k3", "L3"]] = values_df[["k1", "L1", "k2", "L2", "k3", "L3"]].astype(int)
    values_df[rhos] = values_df[rhos].astype(float)

    return SU3_Wigner(values_df)
    
# Class to store the SU(3) Wigner values 
class SU3_Wigner:
    
    def __init__(self, values):

        self.values = values
         
# Funcion to upload 9-SU(3) symbol values from program's output format

def load_9_SU3_file(filename):

    # Open file
    with open(filename + '.txt') as f:
        lines = f.readlines()
        
    # SU3 irreps loaded
    l1, m1 = int(lines[7].split(" ")[0]), int(lines[7].split(" ")[1]) 
    l2, m2 = int(lines[9].split(" ")[0]), int(lines[9].split(" ")[1]) 
    l12, m12 = int(lines[11].split(" ")[0]), int(lines[11].split(" ")[1]) 

    l3, m3 = int(lines[13].split(" ")[0]), int(lines[13].split(" ")[1]) 
    l4, m4 = int(lines[15].split(" ")[0]), int(lines[15].split(" ")[1]) 
    l34, m34 = int(lines[17].split(" ")[0]), int(lines[17].split(" ")[1]) 
    
    l13, m13 = int(lines[19].split(" ")[0]), int(lines[19].split(" ")[1]) 
    l24, m24 = int(lines[21].split(" ")[0]), int(lines[21].split(" ")[1]) 
    l, m = int(lines[23].split(" ")[0]), int(lines[23].split(" ")[1]) 
    

    SU3_9 = [[(l3,m3),(l4,m4),(l34,m34)],
            [(l1,m1),(l2,m2),(l12,m12)],
            [(l13,m12),(l24,m24),(l,m)]]
        
    # Column names of the multiplicities
    cols = list(filter(lambda x: x != '', lines[29].split(" ")))
    cols.append("value")
    cols[5] = cols[5][:5]
    
    # Multiplicities values stored
    values = [list(filter(lambda x: x != '', l.split(" "))) for l in lines[30:]]
    values_df = pd.DataFrame(np.array(values), columns = cols)
    values_df["value"] = values_df["value"].astype(float)
    values_df[["rho1324","rho24", "rho13", "rho1234", "rho34", "rho12"]] = values_df[["rho1324","rho24", "rho13", "rho1234", "rho34", "rho12"]].astype(int)
    
    return SU3_9_symbol(SU3_9, values_df)
 
    
# Class to store the 9-SU(3) values 
class SU3_9_symbol:
    
    def __init__(self, SU3_9_symbol, values):
        
        self.SU3_9_symbol = SU3_9_symbol
        self.values = values
```

```python
# Function to compute the sumation on the rme
def sums(ex_sigma_s):                                           # SUMS FUNCTION sums(ex_sigma_s) CHECKED
    
    ex_sigma = ex_sigma_s[0]+ex_sigma_s[1]
    
    if (ex_sigma_s == "npm"):
        ex_sigma = "np_"

    if (ex_sigma_s[1] == "p"):
        rs = "rho12"
    else: 
        rs = "rho34"


    if (ex_sigma_s[2] == "m"):
        rss = "rho12"
    else: 
        rss = "rho34"

    # Sumations of 9-(l,m) coefficients over barrho 
    sum_total = 0

    for barrho in (0,1):
        wig_coeff = wc.iloc[0]["rho"+str(barrho)]
        #print("wig_",barrho,": ",wc.iloc[0]["rho"+str(barrho)])

        # Sum of 9-(l,m) over rho_sigma with barrho fixed
        sum_ex_sigma = 0

        for rhosigma in set(globals()[ex_sigma][rs]):
            term = float(globals()[ex_sigma][(globals()[ex_sigma]["rho1324"]==barrho) & (globals()[ex_sigma][rs]==rhosigma)]["value"])
            #print("9-(l,m)",term)

            # Sum of 9-(l,m) over rho_sigma_s with rho_sigma fixed
            sum_ex_sigma_s = 0

            for rhosigmas in set(globals()[ex_sigma_s][rss]):
                sum_ex_sigma_s += float(globals()[ex_sigma_s][(globals()[ex_sigma_s]["rho1324"]==rhosigma) & (globals()[ex_sigma_s][rss]==rhosigmas)]["value"])

            #print("sum_ex_sigma_s",sum_ex_sigma_s)
            sum_ex_sigma += term*sum_ex_sigma_s

        #print("sum_ex_sigma",sum_ex_sigma)
        sum_total += wig_coeff*sum_ex_sigma

    return sum_total

# Function to compute the quadrupole reduced matrix elements
def rme(ex_sigma_s, l, m):
    
    spin_factor = 1 
    phase = -1
    
    if (ex_sigma_s[0] == ex_sigma_s[1]):
        spin_factor = 1/2
    if (m == 0):
        phase = 1
        
    return spin_factor*phase*np.sqrt(4*C2(l,m))*sums(ex_sigma_s)        
```

```python
# Reduced matrix elements evaluations
cppm = rme("ppm",25,2)
cppp = rme("ppp",6,0)
cpnm = rme("pnm",34,4)

cnnm = rme("nnm",32,2)
cnnp = rme("nnp",7,0)
cnpm = rme("npm",26,4)
```

```python
# Oscillator parameter and effective charges
b02 = 224**(2/6)
en = effecN(2,90,224)
ep = effecP(2,90,224)

print("b²_0 = ", b02)
print("e_ν = ", en)
print("e_π = ", ep)
```

```python
# Coefficients of beta and one minus beta terms
betacoeff = b02*(ep*cppm + ep*cppp + en*cpnm) 
betaminusonecoeff = b02*(en*cnnm + en*cnnp + ep*cnpm)
print("Resulting equation to solve: \n\n|<3⁻||T²||1⁻>| =",betacoeff,"β +", betaminusonecoeff, "(1-β)  efm²")
```

```python
# Parameters
alpha2Wu = 0.02238
coeffup = -0.44416453
coeffdown = -0.99318207
WeisskopfUnitBE2 = 5.940*10**(-6)*224**(4/3)
coeff31down = -0.77147281
coeff31up = -0.50504751

T2 =(2*3+1)*(alpha2Wu)*(coeff31up**2)*(C2(65,6))
# Reduced matrix element in W.u of the transition
print("From the model\n")
print("\nB(E2; 1⁻ --> 3⁻) = ", (2*3+1)/(2*1+1)*alpha2Wu*coeff31up**2*C2(65,6) ,"  W.u.")
print("\n|<3⁻||T²||1⁻>|² = ", T2 ,"  W.u.")
print("\n|<3⁻||T²||1⁻>|² = ", T2*WeisskopfUnitBE2 ,"  e²b²")
print("\n<3⁻||T²||1⁻> = ", np.sqrt(T2*WeisskopfUnitBE2) ," eb = ", np.sqrt(T2*WeisskopfUnitBE2)*100, " efm²")
```

```python
beta = (np.sqrt(T2*WeisskopfUnitBE2)*100-betaminusonecoeff)/(betacoeff - betaminusonecoeff)
print("β =", beta)
```
